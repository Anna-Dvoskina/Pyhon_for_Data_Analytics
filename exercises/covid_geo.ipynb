```python
import csv
```

```python
MMSA={}
```

```python
with open('covid_geo.csv', 'r') as f:
    line = f.readline()
    print(line)
```

```python
with open('covid_geo.csv', 'r') as f:
    reader = csv.DictReader(f, delimiter=',')
    for row in reader:
        if row['\ufeffMMSA'] in MMSA:
            MMSA[row['\ufeffMMSA']] += 1
        else:
            MMSA[row['\ufeffMMSA']] = 1
MMSA    
```
